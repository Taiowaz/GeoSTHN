```python
import pandas as pd
import os

os.environ["http_proxy"]="http://10.61.1.39:1082"
os.environ["https_proxy"]="http://10.61.1.39:1082"
from tgb.linkproppred.dataset_pyg import PyGLinkPropPredDataset

dataset_name = "thgl-github"
PyGLinkPropPredDataset(name=dataset_name, root="DATA")
```

```python
edgelist = pd.read_csv(
    f"../../DATA/{dataset_name.replace('-','_')}/{dataset_name}_edgelist.csv"
)
nodetype = pd.read_csv(
    f"../../DATA/{dataset_name.replace('-','_')}/{dataset_name}_nodetype.csv"
)

edgemapping = pd.read_csv(
    f"../../DATA/{dataset_name.replace('-','_')}/{dataset_name}_edgemapping.csv"
)
nodemapping = pd.read_csv(
    f"../../DATA/{dataset_name.replace('-','_')}/{dataset_name}_nodemapping.csv"
)
```

```python
subset_edgelist = edgelist.head(50000)
node_min = min(subset_edgelist["head"].min(), subset_edgelist["tail"].min())
node_max = max(subset_edgelist["head"].max(), subset_edgelist["tail"].max())
subset_nodetype = nodetype[nodetype["node_id"].between(node_min, node_max)]
```

```python
os.makedirs(f"../../DATA/{dataset_name.replace('-','_')}_subset", exist_ok=True)
subset_edgelist.to_csv(
    f"../../DATA/{dataset_name.replace('-','_')}_subset/{dataset_name}-subset_edgelist.csv",
    index=False,
)
subset_nodetype.to_csv(
    f"../../DATA/{dataset_name.replace('-','_')}_subset/{dataset_name}-subset_nodetype.csv",
    index=False,
)
edgemapping.to_csv(
    f"../../DATA/{dataset_name.replace('-','_')}_subset/{dataset_name}-subset_edgemapping.csv",
    index=False,
)
nodemapping.to_csv(
    f"../../DATA/{dataset_name.replace('-','_')}_subset/{dataset_name}-subset_nodemapping.csv",
    index=False,
)
```
