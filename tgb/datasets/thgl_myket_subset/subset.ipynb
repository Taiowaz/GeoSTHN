```python
import pandas as pd
import os

os.chdir("/root/LLM-CDHG")
dataset_name = "thgl-myket"
edgelist = pd.read_csv(
    f"tgb/DATA/{dataset_name.replace('-','_')}/{dataset_name}_edgelist.csv"
)
nodetype = pd.read_csv(
    f"tgb/DATA/{dataset_name.replace('-','_')}/{dataset_name}_nodetype.csv"
)
```

```python
subset_edgelist = edgelist.head(10000)
node_min = min(subset_edgelist["head"].min(), subset_edgelist["tail"].min())
node_max = max(subset_edgelist["head"].max(), subset_edgelist["tail"].max())
subset_nodetype = nodetype.head(node_max)
```
